```python
from collections import defaultdict
class Graph:
    def __init__(self, vertices):
        self.V = vertices
        self.graph = defaultdict(list)
        
    def add_edge(self, u, v):
        self.graph[u].append(v)
    
    def DLS(self, src, goal, iteration):
        if src == goal:
            return True
        if iteration <= 0:
            return False
        for i in self.graph[src]:
            if(self.DLS(i, goal, iteration-1)):
                return True
        return False
    
    def IDDFS(self, src, goal, maxDepth):
        for iteration in range(maxDepth):
            if(self.DLS(src, goal, iteration)):
                return True
        return False
    
```

```python
my_graph = Graph(7)
my_graph.add_edge(0, 1)
my_graph.add_edge(0, 2)
my_graph.add_edge(1, 3)
my_graph.add_edge(1, 4)
my_graph.add_edge(2, 5)
my_graph.add_edge(2, 6)

goal = 3
maxDepth = 2
src = 0

if my_graph.IDDFS(src, goal, maxDepth) == True:
    print("reachable")
else:
    print("not reachable")
```
