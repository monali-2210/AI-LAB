## N-queens problem using Genetic Algorithm


```python
import numpy as np
import pandas as pd
maze = np.array([3, 2, 7, 5, 2, 4, 1, 1])
maze2 = np.array([2, 4, 7, 4, 8, 5, 5, 2])
```

```python
def calc_fitness(grid):
    fitness = [7, 6, 5, 4, 3, 2, 1, 0]
    for i in range(8):
        for j in range(i+1, 8):
            if(grid[i] == grid[j]):
                fitness[i]-=1
            if(grid[i] == grid[j] + (j -i) or grid[i] == grid[j] - (j -i)):
                fitness[i]-=1;
    total_fitness = sum(fitness)
    return total_fitness
```

```python
calc_fitness(maze)
```

```python
def mutation(board):
    position=np.random.randint(0,7)
    board[position]=np.random.randint(1,8)
    return board

def cross_over(board1,board2):
    n_board1 = np.copy(board1)
    n_board2 = np.copy(board2)

    crossOverPoint = np.random.randint(0, 7)
    for i in range(crossOverPoint+1):
        swap=n_board1[i]
        n_board1[i]=n_board2[i]
        n_board2[i]=swap
    return n_board1, n_board2
```

```python
max_fitness = calc_fitness(maze)
curr_fitness = 0
children = []

children.append(maze)
children.append(maze2)

while(max_fitness <= 28):
    n_maze, n_maze2 = cross_over(maze, maze2)
    n_maze = mutation(n_maze)
    n_maze2 = mutation(n_maze2)
    
    children.append(n_maze)
    children.append(n_maze2)

    curr_fitness = max(calc_fitness(n_maze), calc_fitness(n_maze2), calc_fitness(maze), calc_fitness(maze2))
    max_fitness = max(max_fitness, curr_fitness)
    print("Max_fitness", calc_fitness(n_maze))
    print("Max_fitness", calc_fitness(n_maze2))
    maze = n_maze
    maze2 = n_maze2
    
```
